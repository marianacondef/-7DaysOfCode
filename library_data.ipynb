# Day 1 - Importando dados

```python
import pandas as pd
```

## Carregando dados 

```python
url_base = 'https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/blob/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/emprestimos-{}{}.csv?raw=true'
anos = range(2010, 2020)
semestres = [1, 2]

data_frames = {}

for ano in anos:
    for semestre in semestres:
        formatted_ano_semestre = f'{ano}{semestre}'
        url = url_base.format(ano, semestre)
        try:
            data_frames[formatted_ano_semestre] = pd.read_csv(url)
            print(f'Successfully loaded data for {formatted_ano_semestre}')
        except Exception as e:
            print(f'Failed to load data for {formatted_ano_semestre}: {e}')


```

## Concatenando os dados

```python
emprestimos = pd.concat(data_frames, ignore_index=True)
emprestimos
```

## Verificando e excluindo duplicatas

```python
emprestimos.value_counts()
```

```python
emprestimos.drop_duplicates()
```

```python
emprestimos.value_counts()
```

## Verificando valores nulos

```python
valores_nulos = emprestimos.isnull().sum()
valores_nulos
```

```python
has_null_values = emprestimos.isnull().values.any()
if has_null_values:
    emprestimos.dropna()
emprestimos
```

## Importando restante dos dados

```python
exemplares = pd.read_parquet('https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet')
exemplares
```

```python
exemplares.value_counts()
```

```python
exemplares.drop_duplicates()
```

```python
has_null_values = exemplares.isnull().values.any()
if has_null_values:
    exemplares.dropna()
exemplares
```

## Merge dos DataSets

```python
dados_emprestimos = emprestimos.merge(exemplares)
dados_emprestimos
```
