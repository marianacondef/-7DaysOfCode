## Day 1 - Importando dados

```python
import pandas as pd
```

### Carregando dados 

```python
url_base = 'Datasets/dados_emprestimos/emprestimos-{}{}.csv'
anos = range(2010, 2020)
semestres = [1, 2]

data_frames = {}

for ano in anos:
    for semestre in semestres:
        formatted_ano_semestre = f'{ano}{semestre}'
        url = url_base.format(ano, semestre)
        data_frames[formatted_ano_semestre] = pd.read_csv(url)
```

### Concatenando os dados

```python
emprestimos = pd.concat(data_frames, ignore_index=True)
emprestimos
```

### Verificando e excluindo duplicatas

```python
emprestimos.drop_duplicates()
```

## Day 2 - Limpeza dos dados

### Organização dos itens por CDU (classificação decimal universal)

```python
CDU = []

for cdu in dados_emprestimos['localizacao']:
    if cdu in range(0,100):
        CDU.append("Generalidades. Ciência e conhecimento")
    elif cdu in range(100,200):
        CDU.append('Filosofia e psicologia')
    elif cdu in range(200,300):
        CDU.append("Religião")
    elif cdu in range(300,400):
        CDU.append("Ciências sociais")
    elif cdu in range(400,500):
        CDU.append("Classe vaga")
    elif cdu in range(500,600):
        CDU.append("Matemática e ciências naturais")
    elif cdu in range(600,700):
        CDU.append("Ciências aplicadas")
    elif cdu in range(700,800):
        CDU.append("Belas artes")
    elif cdu in range(800,900):
        CDU.append("Linguagem. Língua. Linguística")
    else:
        CDU.append("Geografia. Biografia. História")

dados_emprestimos['classificação'] = CDU
dados_emprestimos.tail()
```

### Análise mensal dos empréstimos

Pelo gráfico acima, observa-se que os mêses com maior índice de empréstimos são *março* e *agosto* e os meses com o menor índice *janeiro*, *julho* e *dezembro*, provavelmente ligado às férias estudantis. 

Assim, é razóavel que as férias dos colaboradores sejam organizadas nesses três meses de menor demanda.

Nesses meses, é recomendável realizar atividades administrativas, como inventário e reorganização do sistema de empréstimo, para os colaboradores ativos.

```python
def frequencia(variavel):
    ## Função que checa frequência percentual dependendo da variável categórica

    df = pd.DataFrame(dados_emprestimos[variavel].value_counts())
    df.columns = ['quantidade']
    df['percentual'] = round((df['quantidade'] / df['quantidade'].sum())*100,1)

    return df

frequencia('tipo_vinculo_usuario')
```

A princípio, nota-se uma maioria de **alunos de graduação** como público dos empréstimos das bibliotecas, somando cerca de **78%** do percentual total, seguido por alunos de pós-graduação (**14.6%**). Fica evidente que quaiquer esforços de marketing da biblioteca devem focar nessa categoria. 

```python
frequencia('colecao')
```

A coleção **"Acervo circulante"** é a grande maioria, representando quase 100% do acervo dos empréstimos. Nota-se que coleções como *Obras Raras* e *Literatura de Cordel* não foram emprestadas, criando um alerta para o motivo para isso.  

```python
frequencia('biblioteca')
```

A **Biblioteca Central Zila Mamede** estar em maioria entre os empréstimos condiz com o esperado, considerando que abriga a maioria das obras. Outras bibliotecas, como **Biblioteca Setorial do Núcleo de Ensino Superior do Agreste - NESA - Nova Cruz** não receberam nenhum empréstimo, o que pode ser devido a falta de obras no acervo, falta de investimento ou falta de marketing.

### Plotando boxplot de da distribuição dos empréstimos mensais realizados pelos alunos de graduação  e pós-graduação na coleção do acervo circulante

Função para geração de bloxpot

```python
def boxplot(dataset, x, y, subtitulo):
    
  sns.set_theme(palette='Blues',font_scale=1.3)                    
  plt.figure(figsize=(14,10))  

  ax = sns.boxplot(y=y, x=x, data = dataset, color = '#5B7A8E')
  plt.ylim(0,max(dataset[y])*1.1) 
  plt.xticks(fontsize=12, rotation=30)  
  ax.set(xlabel=None,ylabel=None)
  ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
  ax.set_title("Distribuição dos empréstimos mensais"+"\n",size=16,loc='left',weight='bold')
  ax.text(s=subtitulo,x=-0.5,y=max(dataset[y])*1.11,fontsize=18, ha='left',color='gray')  
  ax=ax
```

### Alunos de graduação

```python
boxplot(graduacao_acervo_circulante, 'ano', 'quantidade', 'Alunos de graduação no acervo circulante')
```

Nota-se que a quantidade empréstimos oscila durante os meses do ano entre os alunos de graduação, que representam também a grande maioria do público de empréstimos da biblioteca. É possível observar um declínio quase contínuo entre 2017 e 2019.

### Alunos de pós-graduação

```python
boxplot(posgraduacao_acervo_circulante, 'ano', 'quantidade', 'Alunos de pós-graduação no acervo circulante')
```

O boxplot dos alunos de pós-graduação mostra um aumento na mediana dos empréstimos de 2010 até 2014, seguido de uma estabilidade até 2017 e uma ligeira diminuição a partir de 2018. A queda a partir de 2018 também está presente no gráfico de graduação, podendo ter uma causa semelhante.

```python
pivot_emprestimos_por_curso = relacao_curso_emprestimos.pivot_table(
        index = 'Curso',
        columns = 'Ano',
        values = 'Quantidade de empréstimos',
        aggfunc= 'sum',
        margins = True,
        margins_name = 'TOTAL',
)
pivot_emprestimos_por_curso
```
