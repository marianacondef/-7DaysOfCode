## Day 1 - Importando dados

```python
import pandas as pd
```

### Carregando dados 

```python
url_base = 'https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/blob/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/emprestimos-{}{}.csv?raw=true'
anos = range(2010, 2020)
semestres = [1, 2]

data_frames = {}

for ano in anos:
    for semestre in semestres:
        formatted_ano_semestre = f'{ano}{semestre}'
        url = url_base.format(ano, semestre)
        data_frames[formatted_ano_semestre] = pd.read_csv(url)
```

### Concatenando os dados

```python
emprestimos = pd.concat(data_frames, ignore_index=True)
emprestimos
```

### Verificando e excluindo duplicatas

```python
emprestimos.value_counts()
```

```python
emprestimos.drop_duplicates()
```

```python
emprestimos.value_counts()
```

### Importando restante dos dados

```python
exemplares = pd.read_parquet('https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet')
exemplares
```

```python
exemplares.value_counts()
```

```python
exemplares.drop_duplicates()
```

### Merge dos DataSets

```python
dados_emprestimos = emprestimos.merge(exemplares)
dados_emprestimos
```

## Day 2 - Limpeza dos dados

### Organização dos itens por CDU (classificação decimal universal)

```python
CDU = []

for cdu in dados_emprestimos['localizacao']:
    if cdu in range(0,100):
        CDU.append("Generalidades. Ciência e conhecimento")
    elif cdu in range(100,200):
        CDU.append('Filosofia e psicologia')
    elif cdu in range(200,300):
        CDU.append("Religião")
    elif cdu in range(300,400):
        CDU.append("Ciências sociais")
    elif cdu in range(400,500):
        CDU.append("Classe vaga")
    elif cdu in range(500,600):
        CDU.append("Matemática e ciências naturais")
    elif cdu in range(600,700):
        CDU.append("Ciências aplicadas")
    elif cdu in range(700,800):
        CDU.append("Belas artes")
    elif cdu in range(800,900):
        CDU.append("Linguagem. Língua. Linguística")
    else:
        CDU.append("Geografia. Biografia. História")

dados_emprestimos['classificação'] = CDU
dados_emprestimos.tail()
```

### Tornando coluna "matricula_ou_siape" em string

```python
dados_emprestimos['matricula_ou_siape'] = str(dados_emprestimos['matricula_ou_siape'])
dados_emprestimos.head()
```

### Deletando coluna "registro_sistema"

```python
dados_emprestimos.drop(columns=['registro_sistema'],inplace=True)
dados_emprestimos.head()
```

### Análise anual de empréstimos

```python
quantidade_vs_data = pd.DataFrame(dados_emprestimos['data_emprestimo'].value_counts().reset_index())
quantidade_vs_data.columns = ['data','quantidade']
quantidade_vs_data['data'] = pd.to_datetime(quantidade_vs_data['data'])
quantidade_vs_data
```

```python
quantidade_vs_data['data'] = pd.to_datetime(quantidade_vs_data['data'])
quantidade_vs_data['ano'] = quantidade_vs_data['data'].dt.year
emprestimos_por_ano = quantidade_vs_data.groupby('ano')['quantidade'].sum().reset_index()
emprestimos_por_ano
```

```python
quantidade_vs_data['data'] = pd.to_datetime(quantidade_vs_data['data'])
quantidade_vs_data['ano'] = quantidade_vs_data['data'].dt.year
emprestimos_por_ano = quantidade_vs_data.groupby('ano')['quantidade'].sum().reset_index()
emprestimos_por_ano
```

```python
def thousands_formatter(x, pos):
    return f'{int(x):,}'.replace(',', '.')

plt.figure(figsize=[12,8])
sns.lineplot(data=emprestimos_por_ano, x='ano',y='quantidade',marker='o')
plt.xlabel(None)
plt.xticks(fontsize=12, rotation=30)
plt.ylabel(None)
plt.title("Quantidade de exemplares emprestados anualmente" + '\n', fontsize=16, weight='bold')
plt.grid(True)
plt.gca().yaxis.set_major_formatter(FuncFormatter(thousands_formatter))
plt.show()
```

Houve uma queda nos empréstimos de 2013 para 2014, após uma alta contínua desde 2010. Desde 2014, a quantidade de empréstimos continuou diminuindo, atingindo maior queda entre 2016 e 2018.

Essa diminuição repentina em 2013 deve ser investigada. Uma das razões para a queda nos empréstimos pode estar relacionada ao aumento da pirataria online de livros, levando a uma menor adesão dos alunos às bibliotecas e meios burocráticos de aquisição dos livros. Outras causas como:
- Diminuição do marketing da biblioteca
- Diminuição do número de alunos
- Não disponibilidade dos livros usados como texto-base

podem também ser investigadas;

### Análise mensal dos empréstimos

```python
quantidade_vs_data['data'] = pd.to_datetime(quantidade_vs_data['data'])
quantidade_vs_data['mês'] = quantidade_vs_data['data'].dt.month
meses = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
quantidade_vs_data['mês'] = quantidade_vs_data['mês'].map(meses)
emprestimos_por_mes = quantidade_vs_data.groupby('mês')['quantidade'].sum()
emprestimos_por_mes = emprestimos_por_mes.reindex(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
emprestimos_por_mes
```

```python

emprestimos_por_mes = pd.DataFrame(emprestimos_por_mes)

def thousands_formatter(x, pos):
    return f'{int(x):,}'.replace(',', '.')

plt.figure(figsize=[12,8])
sns.lineplot(data=emprestimos_por_mes, x='mês',y='quantidade',marker='o')
plt.xlabel(None)
plt.xticks(fontsize=12, rotation=30)
plt.ylabel(None)
plt.title("Quantidade de exemplares emprestados mensalmente" + '\n', fontsize=16, weight='bold')
plt.grid(True)
plt.gca().yaxis.set_major_formatter(FuncFormatter(thousands_formatter))
plt.show()
```

Pelo gráfico acima, observa-se que os mêses com maior índice de empréstimos são *março* e *agosto* e os meses com o menor índice *janeiro*, *julho* e *dezembro*, provavelmente ligado às férias estudantis. 

Assim, é razóavel que as férias dos colaboradores sejam organizadas nesses três meses de menor demanda.

Nesses meses, é recomendável realizar atividades administrativas, como inventário e reorganização do sistema de empréstimo, para os colaboradores ativos. 

### Análise diária dos empréstimos

```python

quantidade_vs_data['data'] = pd.to_datetime(quantidade_vs_data['data'])
quantidade_vs_data['hora'] = quantidade_vs_data['data'].dt.hour
emprestimos_por_hora = quantidade_vs_data.groupby('hora')['quantidade'].sum().reset_index()
emprestimos_por_hora
```

```python


emprestimos_por_hora = emprestimos_por_hora.sort_values(ascending=True,by='quantidade')
plt.figure(figsize=[12,8])
ax = sns.barplot(data=emprestimos_por_hora,y='quantidade',x='hora',
                 palette='Blues',hue='quantidade',dodge=False)   
ax.set(xlabel='Horário',ylabel=None)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
ax.set_title("Quantidade de exemplares emprestados por faixa horária"+"\n",size=16,loc='left',weight='bold')
ax.text(s='Período de 2010 a 2020',x=-0.5,y=224000,fontsize=12, ha='left',color='gray')
ax=ax
     

```

Nota-se que os períodos de pico durante o dia ocorrem entre às 10h e às 16h. Assim, é razoável que as atividades de não-atendimento direto ao público ocorram em horários de menos presença dos alunos. Os horários das 21h em diante são os mais vagos, notando-se também um menor pico no começo da manhã, até às 8h.
